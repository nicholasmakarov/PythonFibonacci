```python
# Author: Nicholas Makarov
# Date 3/9/2023
# Comments: This program uses python to compile the sequence of the first non-negative n even, odd, and Fibonacci numbers in data frame format

import pandas as pd

# Ask the user to enter an integer value greater than 0
n = int(input("Enter an integer value greater than 0: "))

# Use a conditional statement to check if the userâ€™s entry is > than 0
if n > 0:
  # Create a list for each sequence
  even_numbers = [2 * i for i in range(n)]
  odd_numbers = [2 * i + 1 for i in range(n)]
  fibonacci_numbers = []
  a, b = 0, 1
  for i in range(n):
    fibonacci_numbers.append(a)
    a, b = b, a + b

  # Display the results in a data frame format
  data = {"Even": even_numbers, "Odd": odd_numbers, "Fibonacci": fibonacci_numbers}
  df = pd.DataFrame(data)
  print(df)
else:
  print("Invalid input. Please enter an integer value greater than 0.")
```
